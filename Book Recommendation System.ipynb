## Popularity Based Recommendation System 

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_ratings=ratings_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_ratings
```

```python
num_ratings.rename(columns={'Book-Rating':'No. of Book Ratings'},inplace=True)
```

```python
num_ratings
```

```python
avg_ratings=ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
```

```python
avg_ratings
```

```python
avg_ratings.rename(columns={"Book-Rating":"Avg Rating"},inplace=True)
```

```python
avg_ratings
```

```python
popular_df=pd.merge(num_ratings,avg_ratings,on="Book-Title",how="inner")
```

## Collaborative Filtering Based Recommendation System 

```python
x=ratings_with_name.groupby("User-ID").count()["Book-Rating"]>200
```

```python
df=x[x].reset_index()
```

```python
df
```

```python
ratings_with_name
```

```python
filtered=pd.merge(df,ratings_with_name,on="User-ID",how="inner")
```

```python
filtered.drop("Book-Rating_x",axis=1,inplace=True)
```

```python
filtered
```

```python
y=filtered.groupby("Book-Title").count()["Book-Rating_y"]>=50
```

```python
DF=y[y].reset_index()
```

```python
DF
```

```python
final_ratings=pd.merge(filtered,DF,on="Book-Title",how="inner")
```

```python
final_ratings
```

```python
final_ratings.drop("Book-Rating_y_y",axis=1,inplace=True)
```

```python
final_ratings.rename(columns={"Book-Rating_y_x":"Book-Rating"},inplace=True)
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('1984')
```
